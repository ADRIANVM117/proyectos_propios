```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('/content/covid_worldwide.csv')
df.dropna(axis = 0, inplace = True)
df['Total Cases'] = df['Total Cases'].astype(str).str.replace(',', '').astype(int)
df['Total Deaths'] = df['Total Deaths'].astype(str).str.replace(',', '').astype(int)
df['Total Recovered'] = df['Total Recovered'].astype(str).str.replace(',', '').astype(int)
df['Active Cases'] = df['Active Cases'].astype(str).str.replace(',', '').astype(int)
df['Total Test'] = df['Total Test'].astype(str).str.replace(',', '').astype(int)
df['Population'] = df['Population'].astype(str).str.replace(',', '').astype(int)
df
```

```python
sin_usa = df.loc[1:5]
fig, axes = plt.subplots(2, 3, figsize=(20, 12))
sns.barplot(ax=axes[0, 0], data=sin_usa, x='Country', y='Total Deaths', palette = 'Blues_d')
sns.barplot(ax=axes[0, 1], data=sin_usa, x='Country', y='Total Cases', palette = 'Blues_d')
sns.barplot(ax=axes[0, 2], data=sin_usa, x='Country', y='Total Recovered', palette = 'Blues_d')
sns.barplot(ax=axes[1, 0], data=sin_usa, x='Country', y='Total Cases', palette = 'Blues_d')
sns.barplot(ax=axes[1, 1], data=sin_usa, x='Country', y='Total Test', palette = 'Blues_d')
sns.barplot(ax=axes[1, 2], data=sin_usa, x='Country', y='Population', palette = 'Blues_d')
plt.show()
```

```python
df_pinguino = pd.read_csv('/content/penguins_lter.csv')
df_pinguino.isna().sum()
```

```python
df_pinguino['Culmen Length (mm)'].fillna(value = df_pinguino['Culmen Length (mm)'].mean(), inplace = True)
df_pinguino['Culmen Depth (mm)'].fillna(value = df_pinguino['Culmen Depth (mm)'].mean(), inplace = True)
df_pinguino['Flipper Length (mm)'].fillna(value = df_pinguino['Flipper Length (mm)'].mean(), inplace = True)
df_pinguino['Body Mass (g)'].fillna(value = df_pinguino['Body Mass (g)'].mean(), inplace = True)
df_pinguino['Sex'].fillna('Desconocido', inplace = True)
df_eda = df_pinguino.drop(['Comments','Delta 15 N (o/oo)','Delta 13 C (o/oo)'], axis = 1)                                                   
```

```python
plt.figure(figsize = (8,7))
sns.countplot(x = 'Species', data = df_eda,palette = 'Blues')
plt.xticks(rotation=30)
```

```python
# calculamos el tamaÃ±o de versicolor_petal_lenght con len()
numero_de_datos = len(df_eda)
# El numero de intervalos (bins) es la raiz cuadrada del numero de datos
n_bins = np.sqrt(numero_de_datos)
#convertimos n_bins a un numero entero 
n_bins = int(n_bins)
n_bins
#--------------------------------------------------------------------------------------------------------
# Multiples parcelas 
# seleccionar columnas para histogramas
cols = ['Culmen Length (mm)', 'Culmen Depth (mm)','Flipper Length (mm)']
# creamos la figura y el eje
fig, axes = plt.subplots(3, 1, figsize=(8,10))
axes = axes.ravel()  # aplanar la matriz hace mas facil la indexaciÃ³n

for col, ax in zip(cols, axes):
  sns.histplot(data=df_eda[col],kde=True, stat='density', ax=ax, bins =n_bins)
fig.tight_layout()
plt.show()
#------------------------------------------------------------------------------
```

```python
plt.figure(figsize = (6,7))
sns.stripplot(x ='Sex', y = 'Body Mass (g)', data = df_eda, hue = 'Species', palette = 'inferno')
plt.show()
```

- The gentto penguin  have more body mass than others. and the diference is great. the maleÂ´s getto penguin  are bigger than others, aprox +500 grams more.


```python
#swarmplot


fig,axes = plt.subplots(1,2, figsize = (10,8)) 
sns.swarmplot(ax =  axes[0], x ='Sex', y = 'Body Mass (g)', data = df_eda, hue = 'Species', palette ='inferno')
sns.swarmplot(ax = axes[1], x = 'Species', y = 'Culmen Length (mm)', hue = 'Sex', data = df_eda,  palette = 'cool')
plt.xticks(rotation = 30)
plt.show()
```

```python
#boxplot 


cols = ['Body Mass (g)', 'Culmen Length (mm)','Culmen Depth (mm)', 'Flipper Length (mm)']
fig, axes = plt.subplots(4,1, figsize =(14,12))
axes = axes.ravel()  # aplanar la matriz hace mas facil la indexaciÃ³n
for col, ax in zip(cols, axes):
  sns.boxplot(ax = ax, data = df_eda, x = col, y = 'Species', palette = 'inferno')
fig.tight_layout()
plt.show()
```
