```python
mu = adelie['Flipper Length (mm)'].mean()
sigma = np.std(adelie['Flipper Length (mm)'])
print(mu)
print(sigma)
from scipy.stats import norm
# proba ilidad de que el pico mida menos de 190 mm? 
a = norm.cdf(190, mu,sigma)

# probabilidad de que un pico mida mas  de 200 mm? 
b = 1- norm.cdf(200, mu,sigma)
print(b)

# probabilidad de que un pico mida entre 120 mm y 150 mm  
c = norm.cdf(150, mu,sigma) - norm.cdf(120, mu,sigma)
print(c)

# ¿que tamaño sera inferior al 45%? 
d = norm.ppf(.45, mu,sigma)
print(d)  #<--- el 65% de las veces se espera que el tamaño del pico de al menos 189.14 mm 
```

## Toerma del limite central 


```python
import seaborn as sns
import matplotlib.pyplot as plt
numero = len(adelie)
bins = np.sqrt(numero)
n_bins = int(bins)

cols = ['Culmen Length (mm)', 'Culmen Depth (mm)','Flipper Length (mm)', 'Body Mass (g)']
# creamos la figura y el eje
fig, axes = plt.subplots(2, 2, figsize=(12,8))
axes = axes.ravel()  # aplanar la matriz hace mas facil la indexación

for col, ax in zip(cols, axes):
  sns.histplot(data=adelie[col],kde=True, stat='density', ax=ax, bins =n_bins)
fig.tight_layout()
plt.show()
```

```python
# usare la distribución de 'Body Mass (g)', tiene distribución desconocida 
# tomare diferentes muestras y a cada una le sacaremos su media muestral 
medias_muestrales = []
for i in range(1000):
  samples_20 = adelie['Body Mass (g)'].sample(20, replace = True) # tomamos muestra con 20 observaciones, aleatoria con reemplazo 
  media = np.mean(samples_20)
  medias_muestrales.append(media)

#convertimos la lista en una serie para poder graficar su histograma
medias_muestrales_serie = pd.Series(medias_muestrales)
sns.histplot(data = medias_muestrales_serie, kde = True).set(title = 'Distribución de Medias muestrales')
plt.show()
print('\n')

fig, axes = plt.subplots(1,2, figsize= (15,7))
sns.histplot(ax = axes[0], data = medias_muestrales_serie, kde = True).set(title = 'Distribución de Medias muestrales')
sns.histplot(ax = axes[1], data = adelie['Body Mass (g)'], kde = True).set(title = 'Distribución de datos Originales')
plt.show()
```

```python
# la media de la medias muestrales vs la media poblacional 
media_medias_muestrales = np.mean(medias_muestrales)
print('Media de medias muestrales: ', media_medias_muestrales)
print('Media Poblacional: ', np.mean(adelie['Body Mass (g)']))
```

```python
plt.figure(figsize = (20,10))
sns.countplot(x = 'Date Egg', data = df)
plt.xticks(rotation = 30)
plt.show()
```

```python
# manipulación de datos para obtener los nacimientos promedio
a = df['Date Egg'].value_counts()
a_df = pd.DataFrame(a)
a_df['nacimientos'] = a_df['Date Egg']
a_df['nacimientos'].apply(str)
lambda_nacimientos = a_df['nacimientos'].mean()
print('Nacimientos promedio al dia: Lambda', int(lambda_nacimientos))
```

```python
# Import poisson from scipy.stats
from scipy.stats import poisson 
numero = len(a_df)
bins = np.sqrt(numero)
n_bins = int(bins)

# Probabilidad que nazcan 5 pinguinos 
prob_5 = poisson.pmf(5,6)
print(prob_5)
# Probabilidad que nazcan 2 pinguinos 
prob_1 = poisson.pmf(2,6)
print(prob_1)
#probabilidad de que nazcan menos de 5 pinguinos 
prob_5_menos = poisson.cdf(5,6)
print(prob_5_menos)
# Probabilidad de mas de 10 pinguinos
prob_mas_10 = 1 - poisson.cdf(10,6)
print(prob_mas_10)
print('\n')
sns.histplot(data = a_df['nacimientos'], kde = True, bins = n_bins)
plt.show()
```
