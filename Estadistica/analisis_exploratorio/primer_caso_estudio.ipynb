```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# funcion ECDF
def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""
    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, len(x) + 1) / n

    return x, y

# Funcion coeficiente pearson 

def pearson_r(x,y):
  corr_mat = np.corrcoef(x,y)
  
  return corr_mat[0,1]

# Funcion de diferencia de medias 

def diff_of_means(data_1, data_2):
    """Difference in means of two arrays."""

    # The difference of means of data_1, data_2: diff
    diff = np.mean(data_1) - np.mean(data_2)

    return diff

# Funcion bootstraping 

def bootstrap_replicate_1d(data, func):
    """Generate bootstrap replicate of 1D data."""
    bs_sample = np.random.choice(data, len(data))
    return func(bs_sample)

# Funci√≥n bootstrap replicas 

def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates

#  Funcion Bootstraping para una regresion lineal 

def draw_bs_pairs_linreg(x, y, size=1):
    """Perform pairs bootstrap for linear regression."""

    # Set up array of indices to sample from: inds
    inds = np.arange(len(x))

    # Initialize replicates: bs_slope_reps, bs_intercept_reps
    bs_slope_reps = np.empty(size)
    bs_intercept_reps = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_inds = np.random.choice(inds, size=len(inds))
        bs_x, bs_y = x[bs_inds], y[bs_inds]
        bs_slope_reps[i], bs_intercept_reps[i] = np.polyfit(bs_x, bs_y, 1)

    return bs_slope_reps, bs_intercept_reps
```

## Estimaciones de los par√°metros de la profundidad del pico

- Estimar la diferencia de la profundidad media del pico de las muestras de G. scandens de 1975 y 2012 e informar de un intervalo de confianza del 95%.

```python
# Compute the difference of the sample means: mean_diff
mean_diff = diff_of_means(bd_2012, bd_1975)

# Get bootstrap replicates of means
bs_replicates_1975 = draw_bs_reps(bd_1975, np.mean, 10000)
bs_replicates_2012 = draw_bs_reps(bd_2012, np.mean, 10000)

# Compute samples of difference of means: bs_diff_replicates
bs_diff_replicates = bs_replicates_2012 - bs_replicates_1975

# Compute 95% confidence interval: conf_int
conf_int = np.percentile(bs_diff_replicates, [2.5, 97.5])

# Print the results
print('difference of means =', mean_diff, 'mm')
print('95% confidence interval =', conf_int, 'mm')
```

## Prueba de hip√≥tesis: ¬øSon los picos m√°s profundos en 2012?

- Su gr√°fico de la ECDF y la determinaci√≥n del intervalo de confianza dejan bastante claro que los picos de G. scandens en Daphne Major se han vuelto m√°s profundos. Pero, ¬øes posible que este efecto se deba simplemente al azar? En otras palabras, ¬øcu√°l es la probabilidad de que obtengamos la diferencia observada en la profundidad media del pico si las medias fueran iguales?

- Cuidado. La hip√≥tesis que estamos comprobando no es que las profundidades de los picos provengan de la misma distribuci√≥n. Para ello podr√≠amos utilizar una prueba de permutaci√≥n. La hip√≥tesis es que las medias son iguales. Para realizar esta prueba de hip√≥tesis, tenemos que desplazar los dos conjuntos de datos para que tengan la misma media y luego utilizar el muestreo bootstrap para calcular la diferencia de medias.

```python
# Compute mean of combined data set: combined_mean
combined_mean =  np.mean(np.concatenate((bd_1975, bd_2012)))

# Shift the samples
bd_1975_shifted = bd_1975 - np.mean(bd_1975) + combined_mean
bd_2012_shifted = bd_2012 - np.mean(bd_2012) + combined_mean

# Get bootstrap replicates of shifted data sets
bs_replicates_1975 = draw_bs_reps(bd_1975_shifted, np.mean, 10000)
bs_replicates_2012 = draw_bs_reps(bd_2012_shifted, np.mean, 10000)

# Compute replicates of difference of means: bs_diff_replicates
bs_diff_replicates = bs_replicates_2012 - bs_replicates_1975

# Compute the p-value
p = np.sum(bs_diff_replicates >= mean_diff) / len(bs_diff_replicates)

# Print p-value
print('p =', p)
```

El valor p es de 0,0034, lo que sugiere que hay una diferencia estad√≠sticamente significativa. Pero recuerde: ¬°es muy importante saber cu√°n diferentes son! En el ejercicio anterior, obtuviste una diferencia de 0,2 mm entre las medias. Debes combinar esto con la significaci√≥n estad√≠stica. Un cambio de 0,2 mm en 37 a√±os es sustancial para los est√°ndares evolutivos. Si siguiera cambiando a ese ritmo, la profundidad del pico se duplicar√≠a en s√≥lo 400 a√±os.

##vCorrelaci√≥n de los datos de las cr√≠as y de los padres

- En un esfuerzo por cuantificar la correlaci√≥n entre las profundidades de los picos de las cr√≠as y de los padres, nos gustar√≠a calcular estad√≠sticas, como el coeficiente de correlaci√≥n de Pearson, entre los padres y las cr√≠as. Para obtener intervalos de confianza sobre esto, necesitamos hacer un bootstrap de pares.

- Ya has escrito una funci√≥n para hacer bootstrap de pares para obtener estimaciones de los par√°metros derivados de la regresi√≥n lineal. Su tarea en este ejercicio es hacer una nueva funci√≥n con la firma de llamada draw_bs_pairs(x, y, func, size=1) que realice el bootstrap de pares y calcule un √∫nico estad√≠stico en las muestras de pares definidas. La estad√≠stica de inter√©s se calcula llamando a func(bs_x, bs_y). En el siguiente ejercicio, se utilizar√° pearson_r para func.

```python
def draw_bs_pairs(x, y, func, size=1):
    """Perform pairs bootstrap for a single statistic."""

    # Set up array of indices to sample from: inds
    inds = np.arange(len(x))

    # Initialize replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_inds = np.random.choice(inds, size = len(inds))
        bs_x, bs_y = x[bs_inds], y[bs_inds]
        bs_replicates[i] = func(bs_x, bs_y)

    return bs_replicates
```

Calcule el coeficiente de correlaci√≥n de Pearson entre la profundidad del pico de los padres y de los hijos para G. scandens. Haga lo mismo para G. fortis. A continuaci√≥n, utiliza la funci√≥n que escribiste en el √∫ltimo ejercicio para calcular un intervalo de confianza del 95% utilizando pares bootstrap.

## Medir la heredabilidad

- Recuerde que el coeficiente de correlaci√≥n de Pearson es la relaci√≥n entre la covarianza y la media geom√©trica de las varianzas de los dos conjuntos de datos. Es una medida de la correlaci√≥n entre los padres y la descendencia, pero puede no ser la mejor estimaci√≥n de la heredabilidad. Si nos detenemos a pensar, tiene m√°s sentido definir la heredabilidad como el cociente entre la covarianza entre los padres y la descendencia y la varianza de los padres solos. En este ejercicio, se estimar√° la heredabilidad y se realizar√° un c√°lculo bootstrap de pares para obtener el intervalo de confianza del 95%.

- Este ejercicio destaca un punto muy importante. La inferencia estad√≠stica (y el an√°lisis de datos en general) no es una empresa "plug-n-chug". Hay que pensar detenidamente en las preguntas a las que se quiere responder con los datos y analizarlas adecuadamente. Si le interesa saber c√≥mo son de heredables los rasgos, la cantidad que definimos como heredabilidad es m√°s adecuada que la estad√≠stica est√°ndar, el coeficiente de correlaci√≥n de Pearson.

```python
def heritability(parents, offspring):
    """Compute the heritability from parent and offspring samples."""
    covariance_matrix = np.cov(parents, offspring)
    return covariance_matrix[0,1] / covariance_matrix[0,0]

# Compute the heritability
heritability_scandens = heritability(bd_parent_scandens, bd_offspring_scandens)
heritability_fortis = heritability(bd_parent_fortis, bd_offspring_fortis)

# Acquire 1000 bootstrap replicates of heritability
replicates_scandens = draw_bs_pairs(
        bd_parent_scandens, bd_offspring_scandens, heritability, size=1000)
        
replicates_fortis = draw_bs_pairs(
        bd_parent_fortis, bd_offspring_fortis, heritability, size=1000)


# Compute 95% confidence intervals
conf_int_scandens = np.percentile(replicates_scandens, [2.5, 97.5])
conf_int_fortis = np.percentile(replicates_fortis, [2.5, 97.5])

# Print results
print('G. scandens:', heritability_scandens, conf_int_scandens)
print('G. fortis:', heritability_fortis, conf_int_fortis)
```

Aqu√≠ tambi√©n vemos que G. fortis tiene una mayor heredabilidad que G. scandens. Esto sugiere que los rasgos de G. fortis pueden estar fuertemente incorporados a G. scandens por hibridaci√≥n introgresiva.

## ¬øEs heredable la profundidad del pico en G. scandens?

- La heredabilidad de la profundidad del pico en G. scandens parece baja. Podr√≠a ser que esta heredabilidad observada se haya conseguido s√≥lo por casualidad y que la profundidad del pico no sea realmente heredable en la especie. Aqu√≠ pondr√°s a prueba esta hip√≥tesis. Para ello, har√°s una prueba de permutaci√≥n de pares.

```python
# Initialize array of replicates: perm_replicates
perm_replicates = np.empty(100000)

# Draw replicates
for i in range(10000):
    # Permute parent beak depths
    bd_parent_permuted = np.random.permutation(bd_parent_scandens)
    perm_replicates[i] = heritability(bd_parent_permuted, bd_offspring_scandens)


# Compute p-value: p
p = np.sum(perm_replicates >= heritability_scandens) / len(perm_replicates)

# Print the p-value
print('p-val =', p)
```

Se obtiene un valor p de cero, lo que significa que ninguna de las 10.000 r√©plicas de pares de permutaci√≥n que se extrajeron ten√≠a una heredabilidad lo suficientemente alta como para coincidir con la observada. Esto sugiere que la profundidad del pico es heredable en G. scandens, pero no tanto como en G. fortis.
